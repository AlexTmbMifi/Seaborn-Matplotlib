```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np 
import pandas as pd
```

```python
df = pd.read_csv('dm_office_sales.csv')
df
```

```python
# Равномерное распределение данных о зарплате
# преимущество зарплаты лежат в диапазоне 60 - 120 к
sns.rugplot(x='salary', data=df, height=0.5); # height - высота графика
```

```python
# построение гистограммы
#sns.set(style='darkgrid') - метод скоро исчезнет
#plt.grid() - альтернативный метод
# displot может не реагировать на plt.figure() из-за другого механизма отрисовки
sns.set_style(style='darkgrid')
sns.displot(data=df, x='salary', bins=50, kde=True, rug=True); # bins - количество интервалов разбиения данных
# kde - плотность вероятности по зарплате. По значениям Count график kde не осмысленный
# rug - равномерное распределение, сразу три графика на одном
```

```python
# похожий метод displot, но displot - более общий метод
# реагирует на plt.figure()
sns.histplot(data=df, x='salary');
```

```python
# отдельный график плотности вероятности, здесь график осмысленный по оси y
# в displot он просто наносился на график 
sns.kdeplot(data=df, x='salary')
```

```python
# histplot + kde - осмысленный график с двумя осями
fig, ax = plt.subplots(figsize=(6,6))
ax2 = ax.twinx()
sns.histplot(data=df, x='salary', ax=ax)
sns.kdeplot(data=df, x='salary', ax=ax2);

plt.tight_layout()
plt.show()
```

```python
np.random.seed(42)

# randint выдаёт равномерно случайные числа - 
# мы получаем каждое значение возвраста с одинаковой вероятностью
sample_ages = np.random.randint(0,100,200)
```

```python
sample_ages = pd.DataFrame(sample_ages,columns=["age"])
```

```python
sample_ages.head()
```

```python
sns.rugplot(data=sample_ages,x='age')
```

```python
plt.figure(figsize=(12,8))
sns.displot(data=sample_ages,x='age',bins=10,rug=True);
```

```python
plt.figure(figsize=(12,8))
sns.displot(data=sample_ages,x='age',bins=10,rug=True,kde=True)
```

```python
sns.kdeplot(data=sample_ages,x='age', clip=[0,100]) # обрезанный kde
```

```python
sns.kdeplot(data=sample_ages,x='age',bw_adjust=0.1, fill=True); # bw_adjust = bins в гистограмме
# fill-закрашивание данных
```
