```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np 
import pandas as pd
```

```python
# Categorical Plots - отображение статистические метрики для каждой из категорий переменной
# Два вида графика - countplot() и barplot()
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df
```

```python
# Категории с подкатегориями
# вертикальный вариант
fig, ax = plt.subplots(figsize=(12, 6))
sns.boxplot(data=df, x = 'parental level of education', y='math score', hue = 'test preparation course');
plt.legend(loc = (1.1, 0.5));
```

```python
# горизонтальный 
fig, ax = plt.subplots(figsize=(12, 6))
sns.boxplot(data=df, y = 'parental level of education', x='math score', hue = 'test preparation course');
plt.legend(loc = (1.1, 0.5));
```

```python
# График зеркального kde + boxplot = violin plot
sns.violinplot(data=df,  x='math score', hue='parental level of education');
plt.legend(loc=(1.1, 0.5));
```

```python
# совмещение двух половинок kde для разных подкатегории с помощью метода split=True
# важно понимать какие категории и подкатегории
fig, ax = plt.subplots(figsize=(12, 6))
sns.violinplot(data=df, y = 'parental level of education', x='math score', hue = 'test preparation course', split=True);
plt.legend(loc = (1.1, 0.5));
```

```python
# убрали boxplot внутри графика inner = None
# bw = 0.1 - меняем ширину полосы
fig, ax = plt.subplots(figsize=(12, 6))
sns.violinplot(data=df, y = 'parental level of education', x='math score', hue = 'test preparation course', split=True, inner = 'quartile');  #'quartile' - обозначение квартилей, 'stick' - отдельные линии
plt.legend(loc = (1.1, 0.5));
```

```python
# получаем что-то среднее между violinplot и всеми точками на графике
# получаем kde по всему набору точек
sns.swarmplot(data=df, x='math score', size = 2);
# также можно сделать группировку и градиент цвета
# если вылетает ошибка, уменьшаем размер точек или увеличиваем размер графика
```
