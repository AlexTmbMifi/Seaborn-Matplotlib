```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np 
import pandas as pd
```

```python
df = pd.read_csv('dm_office_sales.csv')
df
```

```python
# Ð”Ð¸Ð°Ð³Ñ€Ð°Ð¼Ð¼Ð° Ñ€Ð°ÑÑÐµÑÐ½Ð¸Ñ. Ð•Ñ‘ Ð¼Ð¾Ð¶Ð½Ð¾ Ñ€ÐµÐ´Ð°ÐºÑ‚Ð¸Ñ€Ð¾Ð²Ð°Ñ‚ÑŒ Ñ‡ÐµÑ€ÐµÐ· matplotlib Ñ‚.Ðº. seaborn - Ð½Ð°Ð´ÑÑ‚Ñ€Ð¾Ð¹ÐºÐ° Ð½Ð°Ð´ Ð´Ð°Ð½Ð½Ð¾Ð¹ Ð±Ð¸Ð±Ð»Ð¸Ð¾Ñ‚ÐµÐºÐ¾Ð¹
# Ð·Ð°Ð²Ð¸ÑÐ¸Ð¼Ð¾ÑÑ‚ÑŒ Ð¾Ð±ÑŠÐµÐ¼ Ð¿Ñ€Ð¾Ð´Ð°Ð¶ Ð¾Ñ‚ Ð·Ð°Ñ€Ð¿Ð»Ð°Ñ‚Ñ‹
plt.figure(figsize=(8, 6), dpi = 120) #DPI  â€” ÑÑ‚Ð¾ ÐµÐ´Ð¸Ð½Ð¸Ñ†Ð° Ð¸Ð·Ð¼ÐµÑ€ÐµÐ½Ð¸Ñ Ð¿Ð»Ð¾Ñ‚Ð½Ð¾ÑÑ‚Ð¸ Ñ†Ð¸Ñ„Ñ€Ð¾Ð²Ð¾Ð³Ð¾ Ð¸Ð·Ð¾Ð±Ñ€Ð°Ð¶ÐµÐ½Ð¸Ñ. ÐœÐ¾Ð¶Ð½Ð¾ ÑƒÐ»ÑƒÑ‡ÑˆÐ¸Ñ‚ÑŒ Ñ‡Ñ‘Ñ‚ÐºÐ¾ÑÑ‚ÑŒ Ð¸Ð·Ð¾Ð±Ñ€Ð°Ð¶ÐµÐ½Ð¸Ñ
sns.scatterplot(x = 'salary', y = 'sales', data=df, hue = 'level of education');   # hue - Ð´ÐµÐ»ÐµÐ½Ð¸Ðµ Ð½Ð° Ð³Ñ€ÑƒÐ¿Ð¿Ñ‹,'level of education' - ÐºÐ°Ñ‚ÐµÐ³Ð¾Ñ€Ð¸Ð°Ð»ÑŒÐ½Ð°Ñ Ð¿ÐµÑ€ÐµÐ¼ÐµÐ½Ð½Ð°Ñ
# Ð»ÐµÐ³ÐµÐ½Ð´Ð° ÑÐ¾Ð·Ð´Ð°ÐµÑ‚ÑÑ Ð°Ð²Ñ‚Ð¾Ð¼Ð°Ñ‚Ð¸Ñ‡ÐµÑÐºÐ¸ 
# pallete - Ð¼Ð¾Ð¶ÐµÐ¼ Ð¿Ð¾Ð²Ð»Ð¸ÑÑ‚ÑŒ Ð½Ð° Ð¿Ð°Ð»Ð¸Ñ‚Ñ€Ñƒ Ñ†Ð²ÐµÑ‚Ð¾Ð²
```

```python
# Ð”Ð¸Ð°Ð³Ñ€Ð°Ð¼Ð¼Ð° Ñ€Ð°ÑÑÐµÑÐ½Ð¸Ñ Ñ‡ÐµÑ€ÐµÐ· matplotlib
# Ñ‚Ñ€ÑƒÐ´Ð¾Ð·Ð°Ñ‚Ñ€Ð°Ñ‚ Ð±Ð¾Ð»ÑŒÑˆÐµ, Ð½Ð¾ Ñ‚Ð°Ðº Ñ‚Ð¾Ð¶Ðµ Ð¼Ð¾Ð¶Ð½Ð¾
# alpha - Ð¿Ñ€Ð¾Ð·Ñ€Ð°Ñ‡Ð½Ð¾ÑÑ‚ÑŒ Ñ‚Ð¾Ñ‡ÐµÐº, label - ÑƒÐºÐ°Ð·Ð°Ñ‚ÐµÐ»ÑŒ Ð´Ð°Ð½Ð½Ñ‹Ñ…. ÐÐµÐ¾Ð±Ñ…Ð¾Ð´Ð¸Ð¼Ð¾ Ð´Ð»Ñ Ð¿Ð¾ÑÑ‚Ñ€Ð¾ÐµÐ½Ð¸Ñ Ð»ÐµÐ³ÐµÐ½Ð´Ñ‹
plt.scatter(x= df['salary'].loc[df['level of education'] == 'some college'], y = df['sales'].loc[df['level of education'] == 'some college'], alpha = 0.5, label='some college')
plt.scatter(x= df['salary'].loc[df['level of education'] == 'high school'], y = df['sales'].loc[df['level of education'] == 'high school'], alpha = 0.5, label='high school')
plt.scatter(x= df['salary'].loc[df['level of education'] == "associate's degree"], y = df['sales'].loc[df['level of education'] == "associate's degree"], alpha = 0.5, label="associate's degree")
plt.scatter(x= df['salary'].loc[df['level of education'] == "bachelor's degree"], y = df['sales'].loc[df['level of education'] == "bachelor's degree"], alpha = 0.5, label="bachelor's degree")
plt.scatter(x= df['salary'].loc[df['level of education'] == "master's degree"], y = df['sales'].loc[df['level of education'] == "master's degree"], alpha = 0.5, label="master's degree")

plt.legend(loc='upper left');
```
