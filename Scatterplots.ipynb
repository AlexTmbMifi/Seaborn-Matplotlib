```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np 
import pandas as pd
```

```python
df = pd.read_csv('dm_office_sales.csv')
df
```

```python
# Диаграмма рассеяния. Её можно редактировать через matplotlib т.к. seaborn - надстройка над данной библиотекой
# зависимость объем продаж от зарплаты
plt.figure(figsize=(8, 6), dpi = 120) #DPI  — это единица измерения плотности цифрового изображения. Можно улучшить чёткость изображения
sns.scatterplot(x = 'salary', y = 'sales', data=df, hue = 'level of education');   # hue - деление на группы,'level of education' - категориальная переменная
# легенда создается автоматически 
# pallete - можем повлиять на палитру цветов
```

```python
# Диаграмма рассеяния через matplotlib
# трудозатрат больше, но так тоже можно
# alpha - прозрачность точек, label - указатель данных. Необходимо для построения легенды
plt.scatter(x= df['salary'].loc[df['level of education'] == 'some college'], y = df['sales'].loc[df['level of education'] == 'some college'], alpha = 0.5, label='some college')
plt.scatter(x= df['salary'].loc[df['level of education'] == 'high school'], y = df['sales'].loc[df['level of education'] == 'high school'], alpha = 0.5, label='high school')
plt.scatter(x= df['salary'].loc[df['level of education'] == "associate's degree"], y = df['sales'].loc[df['level of education'] == "associate's degree"], alpha = 0.5, label="associate's degree")
plt.scatter(x= df['salary'].loc[df['level of education'] == "bachelor's degree"], y = df['sales'].loc[df['level of education'] == "bachelor's degree"], alpha = 0.5, label="bachelor's degree")
plt.scatter(x= df['salary'].loc[df['level of education'] == "master's degree"], y = df['sales'].loc[df['level of education'] == "master's degree"], alpha = 0.5, label="master's degree")

plt.legend(loc='upper left');
```

```python
# если в hue содержится непрерывная переменная, то возникает градиент цвета
sns.scatterplot(x = 'salary', y = 'sales', data=df, hue = 'salary');  # hue - деление на группы,'level of education' - категориальная переменная
```

```python
# Увеличиваем размер точек в зависимости от зарплаты
sns.scatterplot(x = 'salary', y = 'sales', data=df, size='salary');
# в size можно передавать и категориальные переменные
```

```python
sns.scatterplot(x = 'salary', y = 'sales', data=df, hue = 'level of education', style = 'level of education', palette='viridis', s = 200);
# s - сделали все точки побольше
# style - поменяли стиль точек
```
